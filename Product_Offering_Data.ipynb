```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
# Load the dataset from CSV
df = pd.read_csv('Product_Offering_Data.csv')

# Display basic information
print("Dataset loaded successfully!")
print(f"\nShape: {df.shape[0]} rows, {df.shape[1]} columns")
```

```python
# Display the first few rows
print("First 10 rows of the dataset:")
df.head(10)
```

```python
# Display all rows to see the complete dataset
print("Complete Dataset:")
df
```

```python
# Check data types
print("Data Types:")
print(df.dtypes)
```

```python
# Get summary statistics
print("Statistical Summary:")
df.describe(include='all')
```

```python
# Check for info about the dataset
print("Dataset Info:")
df.info()
```

```python
# Count missing values per column
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Create a summary DataFrame
missing_summary = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage (%)': missing_percentage.round(2)
})

print("Missing Values Summary:")
print(missing_summary)
print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

```python
# Visualize missing values
print("\nRows with missing values:")
df[df.isnull().any(axis=1)]
```

```python
# Check for complete duplicates
total_duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {total_duplicates}")

# Display the duplicate rows
if total_duplicates > 0:
    print("\nDuplicate rows:")
    print(df[df.duplicated(keep=False)].sort_values('Product_ID'))
```

```python
# Check for duplicates based on Product_ID (primary key)
duplicate_ids = df[df.duplicated(subset=['Product_ID'], keep=False)]
print(f"\nDuplicate Product_IDs count: {duplicate_ids['Product_ID'].nunique()}")
print("\nRows with duplicate Product_IDs:")
duplicate_ids.sort_values('Product_ID')
```

```python
# Check unique values in each categorical column
categorical_columns = ['Product_Type', 'Risk_Level', 'Target_Income_Group']

print("Unique values in categorical columns:")
for col in categorical_columns:
    print(f"\n{col}:")
    print(f"  Unique values: {df[col].unique().tolist()}")
    print(f"  Value counts:")
    print(df[col].value_counts())
```

```python
# Check for whitespace issues or case inconsistencies
print("Checking for whitespace and case inconsistencies:")
for col in categorical_columns:
    # Check for leading/trailing whitespace
    whitespace_issues = df[col].apply(lambda x: x != str(x).strip() if pd.notna(x) else False).sum()
    print(f"\n{col}:")
    print(f"  Whitespace issues: {whitespace_issues}")
    
    # Display unique values with their lengths to spot hidden characters
    unique_vals = df[col].dropna().unique()
    print(f"  Values with lengths: {[(v, len(v)) for v in unique_vals]}")
```

```python
# Check Target_Age_Group column specifically
print("Target_Age_Group Analysis:")
print(f"Unique values: {df['Target_Age_Group'].unique()}")
print(f"Non-null count: {df['Target_Age_Group'].notna().sum()}")
print(f"Null count: {df['Target_Age_Group'].isna().sum()}")
print(f"\nConclusion: Target_Age_Group has 100% missing values - will be dropped.")
```

```python
# Create a copy of the original dataframe for preprocessing
df_cleaned = df.copy()
print(f"Original dataset shape: {df.shape}")
print("Copy created for preprocessing.")
```

```python
# Remove exact duplicates (keeping first occurrence)
print("Before removing duplicates:")
print(f"  Total rows: {len(df_cleaned)}")
print(f"  Duplicate rows: {df_cleaned.duplicated().sum()}")

# Remove duplicates
df_cleaned = df_cleaned.drop_duplicates(keep='first')

print("\nAfter removing duplicates:")
print(f"  Total rows: {len(df_cleaned)}")
print(f"  Rows removed: {len(df) - len(df_cleaned)}")
```

```python
# Verify no duplicates remain
print(f"\nRemaining duplicates: {df_cleaned.duplicated().sum()}")
print(f"Unique Product_IDs: {df_cleaned['Product_ID'].nunique()}")
print(f"Total rows: {len(df_cleaned)}")
```

```python
# Display the cleaned dataset after duplicate removal
print("Dataset after removing duplicates:")
df_cleaned
```

```python
# Check current missing values
print("Current missing values:")
print(df_cleaned.isnull().sum())
```

```python
# Drop the Target_Age_Group column since ALL values are missing (100% null)
# A column with no data provides no useful information

print("Columns before dropping:")
print(df_cleaned.columns.tolist())

# Drop the column
df_cleaned = df_cleaned.drop(columns=['Target_Age_Group'])

print("\nColumns after dropping Target_Age_Group:")
print(df_cleaned.columns.tolist())
```

```python
# Verify missing values after dropping the column
print("Missing values after dropping Target_Age_Group:")
print(df_cleaned.isnull().sum())
print(f"\nTotal missing values: {df_cleaned.isnull().sum().sum()}")
```

```python
# Standardize text columns: strip whitespace
text_columns = ['Product_Name', 'Product_Type', 'Risk_Level', 'Target_Income_Group']

for col in text_columns:
    # Strip leading/trailing whitespace
    df_cleaned[col] = df_cleaned[col].str.strip()
    
print("Text columns standardized (whitespace removed)")
```

```python
# Standardize categorical values to Title Case for consistency
categorical_cols = ['Risk_Level', 'Target_Income_Group']

for col in categorical_cols:
    df_cleaned[col] = df_cleaned[col].str.title()

print("Categorical columns standardized to Title Case")
print(f"\nRisk_Level unique values: {df_cleaned['Risk_Level'].unique().tolist()}")
print(f"Target_Income_Group unique values: {df_cleaned['Target_Income_Group'].unique().tolist()}")
```

```python
# Validate categorical values against expected values
expected_values = {
    'Risk_Level': ['Low', 'Medium', 'High'],
    'Target_Income_Group': ['Low', 'Medium', 'High']
}

print("Validating categorical values:")
for col, expected in expected_values.items():
    actual = df_cleaned[col].unique().tolist()
    invalid = [v for v in actual if v not in expected]
    print(f"\n{col}:")
    print(f"  Expected: {expected}")
    print(f"  Actual: {actual}")
    print(f"  Invalid values: {invalid if invalid else 'None'}")
```

```python
# Convert columns to appropriate data types
print("Current data types:")
print(df_cleaned.dtypes)

# Convert Product_ID to integer
df_cleaned['Product_ID'] = df_cleaned['Product_ID'].astype(int)

# Convert categorical columns to category dtype for efficiency
categorical_columns = ['Product_Type', 'Risk_Level', 'Target_Income_Group']
for col in categorical_columns:
    df_cleaned[col] = df_cleaned[col].astype('category')

print("\nUpdated data types:")
print(df_cleaned.dtypes)
```

```python
# Reset index after removing duplicates
df_cleaned = df_cleaned.reset_index(drop=True)
print("Index reset successfully")
print(f"New index range: 0 to {len(df_cleaned) - 1}")
```

```python
# Final verification
print("=" * 60)
print("DATA PREPROCESSING SUMMARY")
print("=" * 60)

print(f"\n1. SHAPE")
print(f"   Original: {df.shape}")
print(f"   Cleaned:  {df_cleaned.shape}")

print(f"\n2. MISSING VALUES")
print(f"   Original: {df.isnull().sum().sum()}")
print(f"   Cleaned:  {df_cleaned.isnull().sum().sum()}")

print(f"\n3. DUPLICATES")
print(f"   Original: {df.duplicated().sum()}")
print(f"   Cleaned:  {df_cleaned.duplicated().sum()}")

print(f"\n4. UNIQUE PRODUCT IDs")
print(f"   Original: {df['Product_ID'].nunique()}")
print(f"   Cleaned:  {df_cleaned['Product_ID'].nunique()}")

print(f"\n5. COLUMNS")
print(f"   Original: {df.columns.tolist()}")
print(f"   Cleaned:  {df_cleaned.columns.tolist()}")
```

```python
# Display the final cleaned dataset
print("\nFINAL CLEANED DATASET:")
df_cleaned
```

```python
# Final data quality check
print("\nFinal Data Quality Check:")
print(df_cleaned.info())
```

```python
# Display statistics of the cleaned data
print("\nDescriptive Statistics:")
df_cleaned.describe(include='all')
```

```python
# Save the cleaned dataset to a new CSV file
output_filename = 'Product_Offering_Data_Cleaned.csv'
df_cleaned.to_csv(output_filename, index=False)
print(f"Cleaned dataset saved to: {output_filename}")
```
